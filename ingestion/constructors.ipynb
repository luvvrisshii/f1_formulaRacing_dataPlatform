##### Read Data - By Creating DataFrame

```
#Schema
from pyspark.sql.types import StructType, StructField, IntegerType, StringType
```

```
sch = StructType(fields = [
    StructField("constructorId", IntegerType()),
    StructField("constructorRef", StringType()),
    StructField("name", StringType()),
    StructField("nationality", StringType()),
    StructField("url", StringType())
])
```

```
output = filename("2021-03-21")
df = spark.read.schema(sch).json(f"{raw_path}/{output}/constructors.json")
```

#### Columns Manipulation

```
from pyspark.sql.functions import lit
```

```
semi_final_df =df.drop("url").withColumnRenamed("constructorId","constructor_id").withColumnRenamed("constructorRef","constructor_ref")

#ingestion
final_df = current_time_ingestion(semi_final_df)
```

#### Writing to Spark tables in Parquet Format

```
final_df.write.mode("overwrite").format("parquet").saveAsTable("f1_processed.constructors")
```
